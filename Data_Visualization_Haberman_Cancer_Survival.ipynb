```python
#pdf 
sns.FacetGrid(data, hue="status", size=5) \
   .map(sns.distplot, "nodes") \
   .add_legend();
sns.FacetGrid(data, hue="status", size=5) \
   .map(sns.distplot, "year") \
   .add_legend();
sns.FacetGrid(data, hue="status", size=5) \
   .map(sns.distplot, "age") \
   .add_legend();
plt.show();
```

```python
#box plot
sns.boxplot(x='status',y='nodes', data=data)
plt.show()
sns.boxplot(x='status',y='year', data=data)
plt.show()
sns.boxplot(x='status',y='age', data=data)
plt.show()
```

```python
#violin plot
sns.violinplot(x="status", y="nodes", data=data, size=8)
plt.show()
sns.violinplot(x="status", y="year", data=data, size=8)
plt.show()
sns.violinplot(x="status", y="age", data=data, size=8)
plt.show()
```

```python
#multivariate analysis
sns.jointplot(data=data, x="status", y="nodes", kind="reg")
#sns.jointplot(x="petal_length", y="petal_width", data=iris_setosa, kind="kde");
plt.show();
```

```python
sns.jointplot(data=data, x="age", y="nodes",hue="status")
#sns.jointplot(x="petal_length", y="petal_width", data=iris_setosa, kind="kde");
plt.show();
```

```python
#pdf and cdf
counts, bin_edges = np.histogram(data['age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
#print(pdf);
#print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.xlabel('Age')
plt.show()
```

```python
#pdf and cdf
counts, bin_edges = np.histogram(data['year'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
#print(pdf);
#print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.xlabel('Year')
plt.show()
```

```python
#pdf and cdf
counts, bin_edges = np.histogram(data['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
#print(pdf);
#print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.xlabel('Nodes')
plt.show()
```
